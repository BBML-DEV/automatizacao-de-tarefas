```python
#Parte do código responsável por abrir o site e fazer a importação do arquivo excel de forma automatica 

#importar as bibliotecas
import pyautogui # faz a automação do mouse e teclado
import time # controla o tempo do nosso programa
import pyperclip # ela permite a gente copiar e colar com o python

pyautogui.PAUSE = 1
pyautogui.alert("O programa vai começar, não use nada no seu computador enquanto o programa estiver em uso!")
#Comando para entrar no sistema 
pyautogui.hotkey("ctrl", "t")

#escrever na var link o local que o computador deve buscar o arquivo
link = "https://drive.google.com/drive/folders/1wRTFw0sUVBjRr4hW5U9LF7DjLixRyxym"
# copia a informação na var link
pyperclip.copy(link)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
#comando para esperar 5 segundos para executar a página
time.sleep(5)

#comando para selecionar o arquivo e fazer o download da pasta com dados dos clientes
pyautogui.click(x=1119, y=333, clicks=1)
pyautogui.click(x=1712, y=163, clicks=1)
pyautogui.click(x=1516, y=560, clicks=1)
time.sleep(5)
pyautogui.press("enter")
time.sleep(10)
```

```python
#etapa de calcular os indicadores (faturamento e a Quantidade de Produtos)
import pandas as pd
tabela = pd.read_excel(r"C:\Users\pvmar\OneDrive\Imagens\icones notion\Vendas - Dez.xlsx") #caso precise selecionar a aba que o programa deve usar ", sheet_name = número da aba"
display(tabela)
#Calcula a soma totla da quantidade de produtos vendidos e o total de faturamento
faturamento = tabela["Valor Final"].sum()
quantidade_de_produtos = tabela["Quantidade"].sum()
```

```python
#Enviar o e-mail 
# email da diretoria: seugmail+diretoria@gmail.com
pyautogui.hotkey("ctrl", "t")

#usar a url do seu e-mail nessa var link
link = "escreva seu email"
# copia a informação na var link
pyperclip.copy(link)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
#espera um pouco para o e-mail carregar
time.sleep(8)


#escoher o e-mail
pyautogui.click(x=68, y=168)
time.sleep(3)

#usar o email que deseja enviar o relatorio
pyautogui.write("escrevar o email do destinatario")
pyautogui.press("tab")

#preencher o assunto
pyautogui.press("tab")
assunto = "Relatório de vendas."
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")

#preencher o email
pyautogui.press("tab")
texto_email= f"""
Prezados, bom dia
O faturamento foi de: R${faturamento}
A quantidade de produtos foi de: {quantidade_de_produtos}

Abs:
BBML-DEV
"""
pyperclip.copy(texto_email)
pyautogui.hotkey("ctrl", "v")

#enviar o email
pyautogui.hotkey("ctrl", "enter")
```
